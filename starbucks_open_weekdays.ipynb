```python
total = []

for x in r.json()['list']:
    payload['p_sido_cd'] = x['sido_cd']
    temp_r = requests.post(tmp_url, data=payload)
    print(temp_r.json()['list'].__len__())
    total.append(pd.DataFrame(temp_r.json()['list']))
```

```python
df = pd.concat(total)
star_df = df[["s_name","tel","fax","addr","defaultimage","open_dt","doro_address","lat","lot",]].copy()
```

```python
day_df = star_df.copy()
```

```python
date_format = "%Y%m%d"  # 날짜 형식 지정
weekday_values = []  # 요일 값을 저장할 리스트 초기화

# 요일 값에 해당하는 요일명 저장 딕셔너리
days = {
    0: 'Mon', 
    1: 'Tue', 
    2: 'Wed', 
    3: 'Thu', 
    4: 'Fri', 
    5: 'Sat', 
    6: 'Sun'
    }

for i in day_df['open_dt']:
    # 문자열을 datetime 객체로 변환
    date_obj = datetime.strptime(i, date_format)

    print(date_obj)
    weekday = days[date_obj.weekday()]
    weekday_values.append(weekday)
    
    
day_df['open_weekday'] = weekday_values  # 열에 리스트의 값을 할당
```
