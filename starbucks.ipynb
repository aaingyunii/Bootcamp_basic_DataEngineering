```python
# 전국 스타벅스 매장 데이터베이스에 저장
url2 = "https://www.starbucks.co.kr/store/getSidoList.do"
r2 = requests.post(url2)
r2.json()["list"]
```

```python
total = []

for x in r2.json()['list']:
    payload['p_sido_cd'] = x['sido_cd']
    r = requests.post(url, data=payload)
    print(r.json()['list'].__len__())
    total.append(pd.DataFrame(r.json()['list']))
```

```python
df = pd.concat(total)
df2 = df[["s_name","tel","fax","addr","defaultimage","open_dt","doro_address","lat","lot",]].copy()
```
