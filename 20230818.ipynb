## `song_id` 를 통해 곡 정보에 대한 내용을 `return` 하는 함수

```python
import requests
import re
from bs4 import BeautifulSoup

url = "https://www.melon.com/song/detail.htm?songId={}"
head= {"User-Agent":
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36"}

def get_lyrics(id_):
    r = requests.get(url.format(id_), headers=head)
    
    song_txt = BeautifulSoup(r.text, "lxml")
    lyrics = BeautifulSoup(str(song_txt.find("div", class_ ="lyric")).replace("<br/>","\n")).text.strip()
    title, open_dt, type_, _ = [x.text for x in song_txt.find("dl", class_="list").findAll("dd")]
    song_title = song_txt.find("div", class_="song_name").text.split()[-1]
    singer = song_txt.find("div", class_="artist").find("a").find("span").text

    return singer, song_title, lyrics, title, open_dt, type_
```
